Simulating a cosmic ray hit on a PRIMA device from 20230308

```python
from kid_model.mattis_bardeen import *
import numpy as np
import matplotlib.pyplot as plt
from submm.KIDs.res.utils import cardan
```

```python
def iqloop(f, fr, Qr, amp, phi, a):
    '''
    Adapted from nonlinear_iq_ss, but with gain and cable delay removed.
    amp = Qr/Qc
    '''
    deltaf = f - fr
    fg = deltaf / fr
    yg = Qr * fg
    y = np.zeros(f.shape[0])
    #find the roots of the y equation above
    for i in range(0, f.shape[0]):
        y[i] = cardan(4.0, -4.0*yg[i], 1.0, -(yg[i]+a))
    Q_term = amp / np.cos(phi)
    z = 1.0 - Q_term * np.exp(1.0j * phi)/ (1.0 + 2.0j * y)
    return z
```

```python
# device parameters
# L230223.1 MIRA22B AR
# atten = 32dB
# temperature = 150 mK
f0 = 842e6
f_tone = f0
omega_tone = 2*np.pi*f_tone
Qc = 205e3
Qi = 41e3 #internal Q at 150 mK
Qr = 1/(1/Qc+1/Qi)
a = 0.688
phi = -1.205
```

```python
f_tone = f0
f_sweep = np.linspace(f0*(1-10/Qr), f0*(1+10/Qr), 600)
z_sweep = iqloop(f_sweep, f0, Qr, Qr/Qc, phi, a)
frac_df0 = -2.35e-5
z_tone = iqloop(np.array([f_tone*(1+frac_df0)]), f0, Qr, Qr/Qc, phi, a)
f_tone = f_tone*(1+frac_df0)

fig, ax = plt.subplots(figsize=(3,3), dpi=200)
ax.plot(np.real(z_sweep), np.imag(z_sweep), 'b-')
ax.plot(np.real(z_sweep), np.imag(z_sweep), 'b.')
ax.plot(np.real(z_tone), np.imag(z_tone), 'r.')
ax.grid()
```

```python
tau_max = 150e-6
nstar = 120e18#1240e18
E_muon = 0.1e6*eV * 1e-6
Tc = 1.2 #need to refine this and the muon energy values
Delta0 = 1.76*k_B*Tc
V = 17e-18
dNqp0 = E_muon/(1.76*k_B*Tc)
dnqp0 = dNqp0 / V
alpha = 0.8
gamma = 1
N0 = 1.72e10*1e18/eV
T = 0.15
S1 = get_S1(T, Tc, omega_tone)
nth = get_nqp_thermal(T,Tc,N0)
Qqp = 1/(alpha*gamma*S1*nth/(2*N0*Delta0)) #quasiparticle internal Q at 150mK
Q0 = 1/(1/Qi-1/Qqp)

xqp0 = dnqp0/nstar
dt = 1/200e3
t_arr = np.arange(0,500e-6,dt)
dnqp_arr = nstar / ((1+1/xqp0)*np.exp(t_arr/tau_max)-1)
nqp_arr = dnqp_arr + nth
Teff_arr = np.array([get_Teff(nqp,T,Tc,N0) for nqp in nqp_arr])
S2_arr = get_S2(Teff_arr, Tc, omega_tone)
S1_arr = get_S1(Teff_arr, Tc, omega_tone)
dx_arr = alpha*gamma*S2_arr*dnqp_arr/(4*N0*Delta0)
Qqp_arr = 1/(alpha*gamma*S1_arr*nqp_arr/(2*N0*Delta0))

Qr_arr = 1/(1/Qqp_arr + 1/Qc + 1/Q0)
f0_arr = f0*(1-dx_arr)
z_arr = np.array([iqloop(np.array([f_tone]), f0_arr[ii], 
                Qr_arr[ii], Qr_arr[ii]/Qc, phi, a) for ii in range(len(f0_arr))])
```

```python
fig, ax = plt.subplots(figsize=(3,3), dpi=200)
ax.plot(np.real(z_sweep), np.imag(z_sweep))
ax.plot(np.real(z_arr), np.imag(z_arr), 'r.')
ax.grid()
```
